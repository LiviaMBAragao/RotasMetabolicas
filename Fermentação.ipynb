# Fermentação

Mapa das reações da fermentação mostrando os compostos orgânicos mais importantes do processo anaeróbico. As ligações entre os nós representam as reações de conversão enzimática, organizadas em pares substrato-produto através de uma estrutura de dicionário. Os pesos nas conexões indicam quais enzimas são responsáveis por catalizar cada reação da via fermentativa

```python
fermentacao = {
    "piruvato": {
        "lactato": {"label": "lática, lactato-desidrogenase"},
        "NAD+": {"label": "lactato-desidrogenase"},
        "acetaldeído": {"label": "alcóolica, piruvato-descarboxilase"},
        "CO2": {"label": "alcóolica, piruvato-descarboxilase"}
    },
    
    "acetaldeído": {
        "etanol": {"label": "álcool-desidrogenase"},
        "NAD+": {"label": "álcool-desidrogenase"}
    }
}
```

```python
from random import randint
import networkx as nx

SEMENTE_ALEATORIA = randint(1, 100)

print(SEMENTE_ALEATORIA)

NOME_ROTULO = "label"
 
g = nx.DiGraph(fermentacao)
 
posicao = nx.spring_layout(g, seed=SEMENTE_ALEATORIA)

nx.draw(g, with_labels=True, pos=posicao)
 
pesos = nx.get_edge_attributes(g, NOME_ROTULO)

nx.draw_networkx_edge_labels(g, posicao, edge_labels=pesos);
```

```python
from networkx.drawing.nx_agraph import to_agraph
 
fermentacao = to_agraph(g)
 
print(fermentacao.to_string())
 
from IPython.display import Image, display
 
display(Image(fermentacao.draw(format="png", prog="dot")))
```
