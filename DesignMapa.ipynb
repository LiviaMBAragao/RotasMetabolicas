```python
import networkx as nx

from random import randint
SEMENTE_ALEATORIA = randint(1, 1000)

import matplotlib.pyplot as plt

g = nx.DiGraph(mapa_metabolico)

label_aresta = {}

for chave, valor in mapa_metabolico.items(): # substrato = chave e produto = valor
    for valor, enzima in valor.items():
        if isinstance(enzima, dict) and "label" in enzima: 
            label_aresta[(chave, valor)] = enzima["label"]
            
tamanho = plt.figure(figsize=(20, 15))

posicao = nx.spring_layout(g, seed=SEMENTE_ALEATORIA)

labels = nx.draw_networkx_edge_labels(g, posicao, edge_labels = label_aresta, font_color = "magenta")

seta = nx.draw_networkx_edges(g, posicao, arrowstyle = "-")

nx.draw(g, with_labels=True)
```
